# Example 1:Category Classification-20 newsgroups

For example, we want to predict whether a given sentence belongs to the category email, news, sports, computer, etc. In the following example,
we are going to use tf-idi to formulate a feature vector to find the category of documents. 
We will use the data from 20 newsgroup dataset of sklearn.

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.naive_bayes import MultinomialNB 
from sklearn. feature_extraction.text import TfidfTransformer 
from sklearn. feature_extraction.text import CountVectorizer
```

```python
news_train=fetch_20newsgroups(subset='train')
category=news_train.target_names
```

```python
news_train.data[0]
```

```python
#count vectorization for extracting term counts
cv=CountVectorizer()
train_cv=cv.fit_transform(news_train.data)
```

```python
print("dimension of training data:",train_cv.shape)
```

```python
tfidf=TfidfTransformer()
train_tfidf=tfidf.fit_transform(train_cv)
```

```python
#MAKING TESTING DATA


test_data = [
    'We must follow the traffic rules while driving vehicles',
    'Puck is a disk made of rubber',
    'Microwave is an electronic device',
    'Jane is a Christian',
    'Everyone of age 18 and above are eligible to vote'
]
```

```python
classifier=MultinomialNB().fit(train_tfidf,news_train.target)
```

```python
test_cv=cv.transform(test_data)
test_tfidf=tfidf.transform(test_cv)
```

```python
predictions=classifier.predict(test_tfidf)
```

```python
for i in range(len(test_data)):
    print(test_data[i])
    print(category[predictions[i]],'\n')
```

# Example 2:Gender Finder

```python
import random 
import nltk 
from nltk import NaiveBayesClassifier 
from nltk.classify import accuracy 
from nltk.corpus import names
```

```python
nltk.download('names')
```

```python
names.words('male.txt')
```

```python
names.words('female.txt')
```

```python
# list omprehension
male_list= [(name,'male') for name in names.words('male.txt')]
female_list= [(name,'female') for name in names.words('female.txt')]
```

```python
male_list
```

```python
data=male_list+female_list
random.seed(10)
random.shuffle(data)
data
```

```python
test_names=['Treesa','Aparna','Johnson','Josephine','Lovely','Gopi']
```

```python
print(len(data))

#defining training data size 
train_data_size = int(0.8*len(data)) 
print('Training data size: ',train_data_size)
```

```python
def extract_features(name, n):
    last_n_letters = name[-n:]
    return {'feature' :last_n_letters. lower ()}
```

```python
for i in range(1,4):
    print('\nNumber of end letters: ', i)
    features = [(extract_features (name,i), gender) for (name, gender) in data]
    train_data,test_data = features[:train_data_size], features[train_data_size:]
    classifier = NaiveBayesClassifier.train(train_data)
    model_accuracy = accuracy (classifier, test_data)
    print ('Accuracy:',model_accuracy*100, '%') 
    for name in test_names:
        print(name,'-->',classifier.classify(extract_features(name,i)))
```

# Example 3:Sentiment Analysis

```python
from nltk.classify import NaiveBayesClassifier 
from nltk.corpus import movie_reviews 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
nltk.download('stopwords')
nltk.download('movie_reviews')
nltk.download('punkt')
```

```python
#list of all words in 'movie_reviews'
print(movie_reviews.words())
print('Number of words:',len(movie_reviews.words()))
print('categories:',movie_reviews.categories())
```

```python
movie_reviews.fileids('pos')
```

```python
def create_word_features(words):
    #stop word removal
    useful_words=[word for word in words if word not in 
                 stopwords.words('english')]
    my_dict=dict([(word,True) for word in useful_words])
    return my_dict
```

```python
neg_reviews = [] # get the words of negative file ids 
for fileid in movie_reviews.fileids('neg'):
    words = movie_reviews.words(fileid)
    neg_reviews.append((create_word_features(words), "negative"))
    #add the term 'negative' at the end of each set of review words as Labels 
```

```python
print(neg_reviews[0])
print(len(neg_reviews))
```

```python
pos_reviews = [] # get the words of negative file ids 
for fileid in movie_reviews.fileids('pos'):
    words = movie_reviews.words(fileid)
    pos_reviews.append((create_word_features(words), "positive"))
    #add the term 'positive' at the end of each set of review words as Labels 
```

```python
train_set=neg_reviews[:750]+pos_reviews[:750]
test_set=neg_reviews[750:]+pos_reviews[750:]
print(len(train_set),len(test_set))
```

```python
classifier=NaiveBayesClassifier.train(train_set)
```

```python
acc = accuracy(classifier, test_set)
print(acc*100)
```

```python
review_1='''I recently dined at [RestaurantName] and was thoroughly impressed by both the exquisite cuisine and the impeccable service. 
The menu showcased a variety of innovative dishes, blending bold flavors and beautiful presentation. 
The attentive and knowledgeable staff ensured that our meal was a memorable one, providing excellent recommendations and ensuring our satisfaction. 
I can’t recommend [RestaurantName] enough for a fantastic dining experience.'''
```
