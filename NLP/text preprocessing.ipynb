# Tokenization

```python
#punkt is an unsupervised machine learning model provided by nltk library in python to perform tokenization
#triple quotes-docstring,u can use enter and use spaces 
```

```python
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
text=''' Tokenization breaks down a text into individual units.
It can be words, phrases,sentences,or other meaningful elements.'''
word_tokenize(text)
```

```python
sent_tokenize(text) #sentences
```

# stemming

```python
from nltk.stem import PorterStemmer

text='He captures beautiful pictures'
words=word_tokenize(text)
ps=PorterStemmer()
for w in words:
    print('original word:',w)
    print('stemmed word:',ps.stem(w))
```

# lemmatization

```python
#to overcome the disadvantages of stemming we use lemmatization
```

```python
#nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()

for w in words:
    print('original word:',w)
    print('lemmatizeded word:',wnl.lemmatize(w))
```

# stop word removal

```python
#nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
print(stop_words)
```

```python
example_sent='An apple a day keeps the diseases at bay.'
example_sent=example_sent.lower() # to match it with the list of stop words,where all words are lower case
word_tokens=word_tokenize(example_sent)

for word in word_tokens:
    if word not in stop_words:
        print(word,end=' ')
```

# BoW-bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
Sentences=['We are using the Bag of Word model',
          'Bag of Word model is used for extracting the features.']
```

```python
cv=CountVectorizer()
features_text=cv.fit_transform(Sentences).toarray()
print(cv.vocabulary_)
```

# TF-IDF(Term Frequency-Inverse Document Frequency)

```python
#TF- MEASURES HOW OFTEN A TERM APPEARS IN A DOCUMENT.
#IDF-MEASURES HOW UNIQUE OR RARE A TERM IS ACROSS ALL DOCUMENTS.
#Higher the TF-IDF value ,more important the term is in the context
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
corpus=[
    'Machine learning is fascinating.',
    'Natural Language Processing is a subfield of AI.',
    'Machine Learning and Natural Language Processing are usrd in many applications.'
]

vectorizer=TfidfVectorizer()
tfidf_matrix=vectorizer.fit_transform(corpus)
feature_names=vectorizer.get_feature_names_out()
```

```python
df_tfidf=pd.DataFrame(data=tfidf_matrix.toarray(),columns=feature_names)
df_tfidf
```
